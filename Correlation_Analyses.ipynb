Correlation analyses:
    1. Between the metauncertainty parameters for each block of CPDM and the metauncertainty parameter for CRDM
    2. Between the metauncertainty parameters for each block of CPDM and the ambiguity aversion parameter
    3. Between the metauncertainty parameter in CRDM and the ambiguity aversion parameter

```python
from glob import glob
import pandas as pd
import numpy as np
from os.path import join
import math
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SDAN_and_CR_Batch3_Analyses.csv")
df
```

```python
df = df.dropna(axis=0)
df
```

```python
CPDM_b1_mup_Array = df['CPDM b1 mup'].values
CPDM_b2_mup_Array = df['CPDM b2 mup'].values
CPDM_b3_mup_Array = df['CPDM b3 mup'].values
CPDM_b4_mup_Array = df['CPDM b4 mup'].values
CRDM_mup_Array = df['CRDM mup'].values
CRDM_aap_Array = df['CRDM aap'].values
```

Correlation 1

```python
x = CPDM_b1_mup_Array
y = CRDM_mup_Array
plt.scatter(x, y)
plt.title("Metauncertainty CPDM Block 1 vs. Metauncertainty CRDM")
plt.xlabel("Metauncertainty CPDM Block 1")
plt.ylabel("Metauncertainty CRDM")
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)
r = np.corrcoef(x,y)[0,1]
plt.text(np.min(x), np.max(y), "$r$ value: {0:0.3f}".format(r), fontsize=14,
        verticalalignment='top')
plt.show()
```
