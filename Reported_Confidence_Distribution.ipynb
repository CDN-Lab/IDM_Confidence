```python
from glob import glob
import pandas as pd
import numpy as np
from os.path import join
import math

filepath = "/Users/govild2/Documents/CloudResearchTrialDataApprovedFilesBatches1+2/"
good_files = sorted(glob(join(filepath, "idm-v2_2023-*")))

#final arrays of interest 
#to be used in plotting
finalAveragesArrayCRDM = []
finalSDArrayCRDM = []
finalAveragesArrayCDD = []
finalSDArrayCDD= []

#temporary storage arrays per data file
CRDM_single_df_temp_values = []
CDD_single_df_temp_values = []

for i, data in enumerate(good_files):
    df = pd.read_csv(data)

    for index, row in df.iterrows():
        if (((math.isnan(row['crdm_conf_resp.keys'])) == False) & (row['crdm_trial_type'] == 'task')):
            CRDM_single_df_temp_values.append(float(row['crdm_conf_resp.keys']))
        
        if (((math.isnan(row['cdd_conf_resp.keys'])) == False) & (row['cdd_trial_type'] == 'task')):
            CDD_single_df_temp_values.append(float(row['cdd_conf_resp.keys']))
    
    
    finalAveragesArrayCRDM.append(np.mean(CRDM_single_df_temp_values))
    finalSDArrayCRDM.append(np.std(CRDM_single_df_temp_values))

    finalAveragesArrayCDD.append(np.mean(CDD_single_df_temp_values))
    finalSDArrayCDD.append(np.std(CDD_single_df_temp_values))

    CRDM_single_df_temp_values.clear()
    CDD_single_df_temp_values.clear()


#test outputs so far
#print(finalAveragesArrayCRDM)
```
